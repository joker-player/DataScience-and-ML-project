```python
import pandas as pd
```

```python
titanic = pd.read_csv("/content/titanicsurvival.csv")
titanic.shape
```

```python
titanic.head(10)
```

```python
titanic.isnull().sum()
```

```python
titanic['Age'].fillna(method = 'bfill' ,inplace=True)
```

```python
titanic.head(2)
```

```python
dummy = pd.get_dummies(titanic['Sex'])
dummy.head()
```

```python
merge = pd.concat([titanic,dummy],axis='columns')
merge.head()
```

```python
df = merge.drop('Sex',axis='columns')
```

```python
df.head()
```

```python
X = df.drop('Survived',axis= 'columns')
Y = df.Survived
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.2) 
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
model.predict([[3,22.0,7.2500,1,0]])
```
